**SOAL 1**

Diketahui tinggi badan siswa kelas A adalah 100,5 ; 99,7 ; 101 ; 101,3 ; 99,9 ; 98,9 ; 165 ; 102,2 ; 98,7 ; 102,9 ; dan 100. Menggunakan Mean, Median, dan Modus, coba gambarkan pusat distribusi data berikut. Manakah yang paling bagus?

**MEAN**

```python
import matplotlib.pyplot as plt
import math

# Inisialisasi dataset
data = [100.5 , 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]
print(f"Data: {data}")

# Jumlah nilai dalam dataset 
jumlah_nilai = sum(data)

print(f"Jumlah semua nilai: {jumlah_nilai}")

# Jumlah total data dalam dataset 
jumlah_total_data = len(data)
print(f"Jumlah total data: {jumlah_total_data}")

# Jumlah semua nilai dalam dataset 
mean_value = round(sum(data) / jumlah_total_data)
print(f"Mean: {mean_value}")

# Buat Histogram
plt.hist(data, bins = 20, edgecolor = 'black', alpha = 1, color = 'purple')

# Garis Vertikal pada mean
plt.axvline(x = mean_value, color = 'orange', linestyle = 'dashed', linewidth = 1, label = f'rata-rata = {mean_value}')

# Legend
plt.legend(bbox_to_anchor = (1.05, 1), loc = 'upper left')

# menampilkan plot
plt.show()
     
```

**MODUS**

```python
import statistics
import matplotlib.pyplot as plt
from collections import Counter

# Data tinggi badan siswa kelas A
tinggi_badan = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]

# Hitung modus
modus_tinggi_badan = statistics.mode(tinggi_badan)

# Hitung frekuensi distribusi data
frekuensi_data = Counter(tinggi_badan)
tinggi, frekuensi = zip(*frekuensi_data.items())

# Gambarkan histogram
plt.bar(tinggi, frekuensi, width=2, align='edge', edgecolor='black')
plt.title('Histogram Tinggi Badan Siswa Kelas A')
plt.xlabel('Tinggi Badan')
plt.ylabel('Frekuensi')
plt.axvline(modus_tinggi_badan, color='yellow', linestyle='dashed', linewidth=2, label=f'Modus: {modus_tinggi_badan}')
plt.legend()
plt.show()
```
