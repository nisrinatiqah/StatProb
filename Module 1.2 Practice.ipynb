**SOAL 1**

Dalam suatu liga sepak bola, dipilih secara acak 10 pemain dari klub yang berbeda dan diketahui jumlah gol yang mereka raih dalam musim ini adalah: 19, 10, 10, 12, 13, 15, 14, 13, 9, 17. Lakukan analisis terhadap sebaran data dari dataset yang diberikan. Manakah ukuran sebaran yang paling mampu menggambarkan sebaran data dengan baik?

**Range**

```python
# Mencari Range
def calculate_range(skor):
    nilai_tertinggi = max(skor)
    print(f"Nilai tertinggi: {nilai_tertinggi}")

    nilai_terendah = min(skor)
    print(f"Nilai terendah: {nilai_terendah}")

    # Selisih antara nilai tertinggi dan terendah
    data_range = max(skor) - min(skor)
    
    print(f"\n")    
    print(f"Range: {nilai_tertinggi} - {nilai_terendah} = {data_range}")  
    
    return data_range

# Input Data
skor = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
print(f"Data Jumlah Gol : {skor}")
data_range = calculate_range(skor)

import matplotlib.pyplot as plt

# Membuat histogram
plt.hist(skor, bins=30, edgecolor='black', alpha=0.5, color='purple')

y = 1.5 # The y-coordinate where the line will be drawn
x1 = 4/70*1  # The x-coordinate of the start point
x2 = 66/70*1  # The x-coordinate of the end point

# Garis Horizontal
plt.axhline(y=y, xmin=x1, xmax=x2, color='red', linestyle='dashed', linewidth=1, label=f'Range = {calculate_range(data)}')

# Garis vertikal
plt.axvline(x=min(skor), color='green', linestyle='dashed', linewidth=1, label=f'Min Data: {min(skor):.2f}')
plt.axvline(x=max(skor), color='blue', linestyle='dashed', linewidth=1, label=f'Max Data: {max(skor):.2f}')

# Keterangan
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Range')
plt.legend()
plt.show()
```

**Variance**

```python
# Mencari rata-rata
def calculate_mean(skor):
    return sum(skor) / len(skor)

# Mencari variansi
def calculate_variance(skor):
    # Rata-rata 
    mean = calculate_mean(skor)
    print(f"Rata-rata: {mean}")  
    
    # Kuadrat selisih antara setiap nilai data dengan rata-rata
    print(f"Kuadrat selisih setiap nilai data dengan rata-rata:")    
    squared_diff = []
    for x in skor:
        squared_diff_per_x = (x - mean)**2
        squared_diff.append( squared_diff_per_x )
        print(f"({x} - {mean})**2 = {(x - mean)}**2 = {squared_diff_per_x}")    
    
    print(f"Total kuadrat selisih: {sum(squared_diff)}")    
    print(f"Total Jumlah Gol: {len(skor)}")    
    
    # Jumlah kuadrat selisih antara setiap nilai data dengan rata-rata, kemudian dibagi dengan jumlah total data
    variance = sum(squared_diff) / len(skor)
    
    print(f"\n")    
    print(f"Variansi: {sum(squared_diff)} / {len(skor)} = {variance}")    
    
    return variance

# Input Data
skor = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
print(f"Data Jumlah Gol: {skor}")

variance = calculate_variance(skor)
```

**Standard Deviation**

```python
# Fungsi untuk menghitung standar deviasi
def calculate_standard_deviation(skor):
    variance = calculate_variance(skor)
    standard_deviation = variance**0.5
    print(f"\n")   
    print(f"Standard Deviation: {variance}**0.5 = {standard_deviation}")
    
    return standard_deviation

# Inisialisasi dataset
skor = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
print(f"Data Jumlah Gol: {skor}")

standard_deviation = calculate_standard_deviation(skor)
mean = calculate_mean(skor)

import matplotlib.pyplot as plt
import numpy as np

# Membuat histogram
plt.hist(skor, bins=30, edgecolor='black', alpha=0.5, color='yellow')

# Garis vertikal
plt.axvline(x=mean, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean}')
plt.axvline(x=mean - standard_deviation, color='green', linestyle='dashed', linewidth=1, label=f'{mean-standard_deviation:.2f}')
plt.axvline(x=mean + standard_deviation, color='blue', linestyle='dashed', linewidth=1, label=f'{mean+standard_deviation:.2f}')

# Keterangan
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title(f'Standard Deviation = {standard_deviation:.2f}')
plt.legend()
plt.show()
```
